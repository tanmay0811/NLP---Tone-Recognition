```python
k=list()
import glob
path = 'C:/Users/Tanmay/Desktop/Intern work/Tone/Financial_Reports/*.txt'
files = glob.glob(path)
for name in files:
        f=open(name)
        r=f.read()
        r=r.upper()
        k.append(r)     
```

```python
def tone(var):
    n=0
    for nword in var.split():
        if nword in negative:
              n=n+1
    p=0
    for pword in var.split():
        if pword in positive:
              p=p+1
    l=len(var.split())
    t=((p-n)/l)*100
    return t
```
